```python
import pandas as pd
import numpy as np
import json
```

```python
with open('data.json') as json_data:
    df= json.load(json_data)
```

```python
len(df['rows'])
```

```python
df
```

```python
columns=[]
for column in range(0,5):
    columns.append(df['columnHeaders'][column]['name'])

columns
```

```python
rows_day=[]
rows_estimate=[]
rows_views=[]
rows_likes=[]
rows_subs=[]

for row in range(0,len(df['rows'])):
    rows_day.append(df['rows'][row][0])

for row in range(0,len(df['rows'])):
    rows_estimate.append(df['rows'][row][1])

for row in range(0,len(df['rows'])):
    rows_views.append(df['rows'][row][2])

for row in range(0,len(df['rows'])):
    rows_likes.append(df['rows'][row][3])

for row in range(0,len(df['rows'])):
    rows_subs.append(df['rows'][row][4])

print(len(rows_day))
print(len(rows_estimate))
print(len(rows_views))
print(len(rows_likes))
print(len(rows_subs))
```

```python
final_df = pd.DataFrame(
    {'day': rows_day,
     'estimated_minutes_watched': rows_estimate,
     'views': rows_views,
     'likes': rows_likes,
     'subscribers_gained':rows_subs
    })
```

```python
final_df.to_excel("transformed_youtube_dump.xlsx",index=False)
```
